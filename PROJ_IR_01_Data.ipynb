```python
# ================================
# PROJ_IR_01_Data.ipynb
# Data Loading and Preparation
# Air University Islamabad
# ================================

# Cell 1 — Install
!pip install datasets -q
print("✅ Done!")

# Cell 2 — Load Dataset
from google.colab import userdata
from huggingface_hub import login
from datasets import load_dataset

login(token=userdata.get('HF_TOKEN'))

print("Loading IL-PCSR dataset...")
ds_queries    = load_dataset(
    "Exploration-Lab/IL-PCSR",
    name="queries")
ds_statutes   = load_dataset(
    "Exploration-Lab/IL-PCSR",
    name="statutes")
ds_precedents = load_dataset(
    "Exploration-Lab/IL-PCSR",
    name="precedents")
print("✅ Dataset loaded!")

# Cell 3 — Prepare
statute_texts   = [s['text']
    for s in ds_statutes['statute_candidates']]
statute_ids     = [s['id']
    for s in ds_statutes['statute_candidates']]
precedent_texts = [p['text']
    for p in ds_precedents['precedent_candidates']]
precedent_ids   = [p['id']
    for p in ds_precedents['precedent_candidates']]
query_texts     = [q['text']
    for q in ds_queries['test_queries']]
query_ids       = [q['id']
    for q in ds_queries['test_queries']]
gold_statutes   = [q['relevant_statute_ids']
    for q in ds_queries['test_queries']]
gold_precedents = [q['relevant_precedent_ids']
    for q in ds_queries['test_queries']]

print(f"✅ Statutes     : {len(statute_texts)}")
print(f"✅ Precedents   : {len(precedent_texts)}")
print(f"✅ Test queries : {len(query_texts)}")

# Cell 4 — Save immediately
import pickle

with open('processed_data.pkl', 'wb') as f:
    pickle.dump({
        'statute_texts'   : statute_texts,
        'statute_ids'     : statute_ids,
        'precedent_texts' : precedent_texts,
        'precedent_ids'   : precedent_ids,
        'query_texts'     : query_texts,
        'query_ids'       : query_ids,
        'gold_statutes'   : gold_statutes,
        'gold_precedents' : gold_precedents,
    }, f)

print("✅ Saved!")

# Cell 5 — Download
from google.colab import files
files.download('processed_data.pkl')
print("✅ Downloaded! Share with team!")
```
